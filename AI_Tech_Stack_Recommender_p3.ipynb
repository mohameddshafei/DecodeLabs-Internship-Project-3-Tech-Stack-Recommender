# AI Project 3: Tech Stack Recommendation System

## Objective

The objective of this project is to build a simple recommendation system that takes user skills or interests as input, compares them with different technology-related career roles, and recommends the most suitable roles based on similarity.

This project uses recommendation logic, text matching, TF-IDF, and cosine similarity to suggest the top career paths for the user.

## Dataset Description

I created this dataset on my own to showcase my project.
The dataset contains different technology-related job roles and the skills, tools, and descriptions connected to each role.

Columns used:
- role: The career path or job role
- skills: The main skills required for the role
- tools_platforms: Tools and platforms commonly used in the role
- description: A short explanation of the role

## 1- Imports and Inspecting Dataset

```python
#imports
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#loading dataset

df = pd.read_csv("raw_skills.csv")
df = df[["role", "skills", "tools_platforms", "description"]]
```

```python
#checking the first 5 rows
df.head()
```

```python
#checking the information of the dataset
df.info()


#checking missing values
print("-"*50)
df.isnull().sum()
```

## 2- Text Preparation

To compare the user's interests with the roles, the important text columns are combined into one column. This gives the recommendation system more information to compare against.

```python
# Combining the important columns that will affect the recommendations
df["combined_text"] = df["skills"] + " " + df["tools_platforms"] + " " + df["description"]
df[["role", "combined_text"]].head()
```

## 3- User Input

The user enters at least three skills or interests. These inputs will be used to find the closest matching career roles.

```python
#use input
skill1 = input("Enter your first skill or interest: ").lower()
skill2 = input("Enter your second skill or interest: ").lower()
skill3 = input("Enter your third skill or interest: ").lower()

# Combining skills into one sentence
user_input = skill1 + " " + skill2 + " " + skill3

print("Your skills/interests:", user_input)
```

## 4- TF-IDF Vectorization

TF-IDF converts text into numerical values so the system can compare the user's input with each role in the dataset.

```python
# Turn text into numbers so the recommendation system can compare the user's skills with each role.
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df["combined_text"])
user_vector = vectorizer.transform([user_input])
```

## 5- Cosine Similarity

Cosine similarity measures how close the user's input is to each role in the dataset. A higher similarity score means a better match.

```python
#Calculating the scores
similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

#Adding the scores to the dataset
df["similarity_score"] = similarity_scores[0]
```

## 6- Sorting and Recommendation
Sorting the roles from highest similarity score to lowest and recommending the top 3 matches.

```python
#sorting,recommending and showing the table with similarity score
recommendations = df.sort_values(by="similarity_score", ascending=False).head(3)
recommendations[["role", "skills", "tools_platforms", "similarity_score"]]
```

## 7- Displaying Results

```python
#Showing results
print("Top 3 Recommended Career Paths:\n")

for index, row in recommendations.iterrows():
    print("Role:", row["role"])
    print("Skills:", row["skills"])
    print("Tools/Platforms:", row["tools_platforms"])
    print("Similarity Score:", round(row["similarity_score"], 2))
    print("-" * 50)
```

## 8- Conclusion

This project successfully built a simple recommendation system using user skills and career interests. The system compares the user's input with different technology roles and recommends the top matching career paths based on similarity scores.

Through this project, I learned how recommendation systems can use text-based matching, TF-IDF, and cosine similarity to suggest relevant options. The system is simple, but it shows the core idea behind recommendation logic.

## 9- The Final System
This final section combines all previous steps into one complete runnable recommendation system.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv("raw_skills.csv")

df["combined_text"] = df["skills"] + " " + df["tools_platforms"] + " " + df["description"]

skill1 = input("Enter your first skill or interest: ").lower()
skill2 = input("Enter your second skill or interest: ").lower()
skill3 = input("Enter your third skill or interest: ").lower()

user_input = skill1 + " " + skill2 + " " + skill3

print("Your skills/interests:", user_input)

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df["combined_text"])
user_vector = vectorizer.transform([user_input])

#Calculating the scores
similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

#Adding the scores to the dataset
df["similarity_score"] = similarity_scores[0]

recommendations = df.sort_values(by="similarity_score", ascending=False).head(3)
recommendations[["role", "skills", "tools_platforms", "similarity_score"]]

print("Top 3 Recommended Career Paths:\n")

for index, row in recommendations.iterrows():
    print("Role:", row["role"])
    print("Skills:", row["skills"])
    print("Tools/Platforms:", row["tools_platforms"])
    print("Similarity Score:", round(row["similarity_score"], 2))
    print("-" * 50)
```
